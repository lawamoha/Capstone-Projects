```python
display(dbutils.fs.ls("/mnt/retaildata"))
```

```python
df=spark.read.format("parquet").option("header","true").option("inferSchema","true").load("/mnt/retaildata/retail.parquet")
```

```python
display(df)
```

```python
#remove nulls from information
from pyspark.sql.functions import col, upper, trim
df1 = df.filter((col('TransactionDate').isNotNull()) & (col('customerid').isNotNull()) & (col('transactionid').isNotNull()))
display (df1)
```

```python
display(df.count())
display(df1.count())
```

```python
#PaymentType
#storeregion
#deviceused
from pyspark.sql.functions import upper, trim, col
df2 = df1.withColumn("paymenttype", upper(trim(col("paymenttype"))))\
.withColumn("storeregion", upper(trim(col("storeregion"))))\
.withColumn("deviceused", upper(trim(col("deviceused"))))

```

```python
df2.show(30)
```

```python
#TransactionDate
#Quantity
#Amount
#Discount
from pyspark.sql.functions import col
df3 = df2.withColumn("TransactionDate",(col("TransactionDate").cast("timestamp")))\
.withColumn("Quantity",(col("Quantity").cast("int")))\
.withColumn("Amount",(col("Amount").cast("float")))\
.withColumn("Discount", (col("Discount").cast("float")))
```

```python
display(df3)
```

```python
df4 = df3.filter((col("quantity") > 0) & (col("amount") > 0) & (col("discount") <= col("amount")))
```

```python
display(df4)
```

```python
# Rename a column in a DataFrame
df5 = df4.withColumnRenamed("paymenttype", "PaymentType")\
.withColumnRenamed("storeregion", "StoreRegion")\
.withColumnRenamed("deviceused", "DeviceUsed")
display(df5)
```

```python
df6 = df5.dropDuplicates(['TransactionID'])
display(df6)
```

```python
df6.count()
df6.write.mode("overwrite").format("parquet").save("/mnt/retaildata/silver")
df_silver = spark.read.parquet("/mnt/retaildata/silver")
display(df_silver)
```

```python
df_silver.createOrReplaceTempView("retail_data")
```

```python
%sql
select date(transactionDate) as Date , ROUND(SUM(Amount-discount),2) As Revenue_per_day
from retail_data
group by transactionDate
order by Revenue_per_day desc
```

```python
%sql
select date(TransactionDate),sum(amount) as total_revenue,count(distinct TransactionID) as total_purchase 
from retail_data
group by date(TransactionDate)
order by date(TransactionDate)
```

```python
%sql
select paymenttype as Method_of_payment, ROUND(SUM(Amount),2) as Total_revenue
from retail_data
group by paymenttype
order by SUM(Amount) desc
```

```python
%sql
select storelocation, sum(amount) as total_revenue
from retail_data
group by storelocation
```

```python
%sql
select Customerloyaltylevel, SUM(amount)
from retail_data
group by Customerloyaltylevel
```
